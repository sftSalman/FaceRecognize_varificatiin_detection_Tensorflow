```python
# triplet loss 
def triplet_loss(y_true, y_pred, alpha = 0.2):
  an_image , pos_image, neg_image = y_pred[0],y_pred[1],y_pred[2]
  post_dis = tf.reduce_sum(tf.square(tf.subtract(an_image,pos_image)),axis=1)
  neg_dist =  tf.reduce_sum(tf.square(tf.subtract(an_image,neg_image)),axis=1)
  basic_loss = tf.maximum(tf.add(tf.subtract(pos_dis,neg_dis),alpha),0)
  loss = tf.reduce_sum(basic_loss)

  return loss
                                              
```
